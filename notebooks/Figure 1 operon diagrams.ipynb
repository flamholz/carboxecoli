```python
%matplotlib inline

import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib

from matplotlib import pyplot as plt
from rubiscos import PCC7942
```

```python
g = sns.color_palette('Greens', desat=0.8)
o = sns.color_palette('Oranges', desat=0.8)
p = sns.color_palette('Purples', desat=0.8)
sns.palplot(g)
sns.palplot(o)
sns.palplot(p)
```

```python
unknown_cat = 'unk'
csome_shell_cat = 'shell'
csome_interior_cat = 'interior'
ci_trans_cat = 'ci_transport'
rubisco_proteostasis_cat = 'rubisco_proteostasis'
rubisco_cat = 'rub'

cat_colors = {
    unknown_cat: 'grey',
    rubisco_cat: g[-1],
    rubisco_proteostasis_cat: g[-3],
    ci_trans_cat: o[-2],
    csome_interior_cat: p[-1],
    csome_shell_cat: p[-1],
}
```

```python
# Colors for carboxylation and oxygenation
carb_color = cat_colors[rubisco_cat]  # same as rubisco
ox_color = 'grey'

# CO2 and O2 concentrations in uM
# CO2 concentrations range from ~10 uM to 1000 uM. 
# Henry's law at 25 C is ~15 uM for CO2
co2_concs = np.logspace(0.9, 3.1, 200)

# Henry's law at 25 C is ~270 uM for O2
o2_conc = 270

# Calculate rates for the cyanobacterium S. elongatus PCC 7942.
vc_7942, vo_7942 = PCC7942.vs(co2_concs, o2_conc)

# Net rate of carboxylation is carboxylation less 1/2 oxygenation
# (assuming C2 photorespiratory pathway, as in land plants)
net_vc_7942 = vc_7942 - vo_7942/2
```

```python
fig = plt.figure(figsize=(5,5))

plt.xscale('log')
plt.yscale('log')
plt.plot(co2_concs, vc_7942, lw=3, c=carb_color)
plt.plot(co2_concs, vo_7942, lw=3, c=ox_color)


plt.axvline(15, ls='--', color='grey')
plt.axvspan(100, 500, color='grey', alpha=0.4, zorder=-1)

#plt.xlabel('Aqueous [$CO_2$] ($\mu M$)', fontsize=20)
#plt.ylabel('Per Active Site Rate ($s^{-1}$)', fontsize=20)

plt.xlim(8, 1200)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.tight_layout()

plt.savefig('fig1/fig1b.png', dpi=600)
plt.show()
```

```python
# Data on the genes from H. neapolitanus from Desmarais 2019, Nat. Micro. 
orf_df = pd.read_csv('hnea_two_operons.csv', index_col=0)
orf_df
```

```python
# Operon diagram for the two CCM operons that we cloned into plasmids
fig,ax = plt.subplots(figsize=(10, 3.5), frameon=False)

plt.ylim(0, 0.4)

# Arrow parameters
width = 0.05
headwidth = 0.085
height = 1e-5
head_scale = 0.2
len_per_bp = 1e-4

region = (983594-12000,986794+4700)
region_start, region_end = region
region_length = region_end - region_start
length_longest_axis = (max(region)-min(region))*len_per_bp
y = 0.2

plt.axhline(y, lw=3, c='k', zorder=-1)
genes_in_region = orf_df[(orf_df.start_pos >= region_start) & 
                         (orf_df.end_pos <= region_end)]

for gene_idx in genes_in_region.index:
    gene = genes_in_region.loc[gene_idx]
    gene_length = gene.length

    gene_name = gene.short_name
    gene_cat = gene['cat']
    
    c = cat_colors.get(gene_cat)
    length = gene_length * len_per_bp
    head_length = length*head_scale
    if length > 0.03:
        head_length = 0.03

    gene_start = gene.start_pos
    gene_end = gene.end_pos
    gene_strand = gene.strand

    if gene_strand == '+':
        start = (gene_start - region_start)*len_per_bp
        genePic = matplotlib.patches.FancyArrow(
            start, y, length, 0,
            width=width, length_includes_head=True,
            head_width=headwidth, head_length=head_length, fill=True, facecolor=c)
        ax.add_artist(genePic)
    else:
        start = (gene_end - region_start) * len_per_bp
        genePic = matplotlib.patches.FancyArrow(
            start, y, -length, 0,
            width=width, length_includes_head=True,
            head_width=headwidth,
            head_length=head_length, fill=True,facecolor=c)
        ax.add_artist(genePic)

plt.xlim(0,length_longest_axis)

ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)
sns.despine(top=True, right=True, bottom=True, left=True)
plt.tight_layout()
plt.gca().invert_xaxis()
plt.savefig('fig1/fig1c.png', dpi=600)

plt.show()
```
