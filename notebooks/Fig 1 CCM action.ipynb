```python
%matplotlib inline

import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

from rubiscos import PCC7942
```

```python
qual_palette = sns.color_palette('Paired', n_colors=10, desat=0.9)
sns.palplot(qual_palette)
carb_color = qual_palette[3]
ox_color = qual_palette[4]

co2_concs = np.logspace(0.9, 3.1, 200) # uM
o2_conc = 270 # uM

# Cyanobacterial Rubisco
vc_7942, vo_7942 = PCC7942.vs(co2_concs, o2_conc)
net_vc_7942 = vc_7942 - vo_7942/2
```

```python
fig = plt.figure(figsize=(10,10))

plt.xscale('log')
plt.yscale('log')
plt.plot(co2_concs, vc_7942, lw=3, c=carb_color)
plt.plot(co2_concs, vo_7942, lw=3, c=ox_color)


plt.axvline(15, ls='--', color='grey')
plt.axvspan(100, 500, color='grey', alpha=0.4, zorder=-1)

plt.xlabel('Aqueous [$CO_2$] ($\mu M$)', fontsize=26)
plt.ylabel('Per Active Site Rate ($s^{-1}$)', fontsize=26)
plt.xlim(8, 1200)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.tight_layout()
plt.savefig('fig1/ccm_action.png', dpi=600)

plt.show()
```
