```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
fname = '/Users/abrahamflamholz/BDrive/Research/Savage/CarboxeColi/phenotyping/062119_CCMB1_half_ab_test/062119_CCMB1_half_ab_test.xlsx'
data = pd.read_excel(fname)
data['od'] = 5*data['od_5xd']
data.head()
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.barplot(data=data, x='short_name', y='od', hue='co2', palette='Set2', ax=ax)
plt.xticks(fontsize=26)
plt.yticks(fontsize=26)
plt.xlabel('')
plt.ylabel('Mean OD600 @ 6 days', fontsize=26)
plt.legend(fontsize=18)
plt.show()
```
