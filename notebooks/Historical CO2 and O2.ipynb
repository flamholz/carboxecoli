```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns

from matplotlib import pyplot as plt
```

```python
concs_df = pd.read_excel('GEOCARBdata.xlsx', sheet_name='all_data')
```

```python
figure = plt.figure(figsize=(20,5))

pal = sns.color_palette()
plt.yscale('log')

a = plt.plot(concs_df.TIME_MA, concs_df.RO2, c='k', lw=3, label='O$_2$')
plt.ylim(5e-4, 1e3)
plt.xlim(-4000, 0)
plt.yticks(fontsize=18, color='k')
plt.xticks(fontsize=18)
plt.ylabel('Atmospheric O$_2$\n (% of PAL)', fontsize=22, color='k')
plt.xlabel('Millions of Years Ago', fontsize=22)

plt.twinx()
plt.yscale('log')

b = plt.plot(concs_df.TIME_MA, concs_df.RCO2, c='r', lw=3, label='CO$_2$')
plt.ylim(1e0, 1e3)
plt.yticks(fontsize=18, color='r')
plt.ylabel('Atmospheric CO$_2$\n (Relative to PAL)', fontsize=22, color='r')

lines = a+b
labels = [l.get_label() for l in lines]
#plt.legend(lines, labels, fontsize=18, loc=2)

plt.show()
```
