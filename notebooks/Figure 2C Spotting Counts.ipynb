```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib.colors import LogNorm
from matplotlib import cm
import dabest

from matplotlib.ticker import StrMethodFormatter, NullFormatter
```

```python
# Make a color palette for the swarmplot that colors replicates differently.
greys = sns.color_palette('Greys', n_colors=8)
purples = sns.color_palette('Purples', n_colors=8)
plt.rc('axes', linewidth=2)
plt.rc('xtick.major', size=8)
plt.rc('xtick.major', width=1.25)
plt.rc('ytick.major', size=8)
plt.rc('ytick.major', width=1.25)
```

```python
fname = '/Users/abrahamflamholz/BDrive/Research/Savage/CarboxeColi/phenotyping/031819 CCMB1 anaerobic spotting counts/031819_CCMB1_anaerobic_colony_counts.xlsx'
```

```python
cts_df = pd.read_excel(fname)
cts_df['log_CFU'] = np.log10(cts_df['CFU_od_ml'])
cts_df.cond.unique()
```

```python
# Load the above data into `dabest`.
cols = ['short_name', 'colony', 'cond', 'CFU_od_ml', 'log_CFU']
sub_df = cts_df[cts_df.short_name == 'CCMB1+pFE1A'][cols].copy()
sub_df
```

```python
conds = ['ambient_no3', '10%_no3', 'anaerobic_no3']
counts = sub_df.groupby('cond').count().loc[conds]['log_CFU']
means = sub_df.groupby('cond').mean().loc[conds]['log_CFU']

print(counts.values)
print(np.power(10, means.values))

fig, ax = plt.subplots(figsize=(5,8))
#plt.bar(np.arange(3), means.values, color='grey')
#plt.yscale('log')
ax.yaxis.set_minor_formatter(NullFormatter())
sns.boxplot(data=sub_df, x='cond', y='log_CFU', color=greys[1], order=conds, whis=np.inf, width=0.45)
sns.stripplot(data=sub_df, x='cond', y='log_CFU', color=greys[-2],
              order=conds, size=9, marker='D', edgecolor='black') 

plt.ylim(7.7, 9.3)
plt.yticks([8, 9], ['$10^{%d}$' % i for i in np.arange(8, 10, 1)], fontsize=16)
plt.ylabel('Colony Forming Units (CFU/OD/mL)', fontsize=20)
plt.xticks(np.arange(3), ['Ambient', '10% CO$_2$', 'Anaerobic'], fontsize=18)
plt.xlabel('')

sns.despine(fig)
plt.tight_layout()
plt.savefig('fig2/counts_anaerobic.png', dpi=300)
plt.show()
```
