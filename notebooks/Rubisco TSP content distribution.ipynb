```python
%matplotlib inline
import numpy as np
import seaborn as sns
import pandas as pd

from matplotlib import pyplot as plt
```

```python
df = pd.read_excel('data/galmes2014_tsp_data.xlsx', index_col=0)

# limiting to taxa that are primarily autotrophic (e.g. not carnivorous)
# and have no CCM for sure (e.g. not bryophytes, some of which have pyrenoid).
relevant_taxa = ['Ferns', 'Gymnosperms', 'Basal angiosperm',
       'C3 plants from low stress environments',
       'C3 plants from environments with high osmotic and heat stress']
ccm_taxa = ['CAM plants', 'Bryophytes']

sub_df = df[df.taxonomic_note.isin(relevant_taxa)]
ccm_df = df[df.taxonomic_note.isin(ccm_taxa)]
```

```python
colors = sns.color_palette('dark')
figure = plt.figure(figsize=(10,10))
plt.hist(sub_df.pct_tsp, color=colors[0], bins=10)
plt.hist(ccm_df.pct_tsp, color=colors[1])

xs = np.arange(0, 41, 10)
plt.xticks(xs, ['%d%%' % x for x in xs], fontsize=20)
plt.yticks(np.arange(1, 4), fontsize=20)

plt.xlabel('[Rubisco]/[TSP] (%)', fontsize=24)
plt.ylabel('Number of Measurements', fontsize=24)
plt.show()
```
